```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import sys
import statistics
import time
from tqdm import tqdm
sys.path.append(os.path.dirname(os.getcwd()))

from problems.problem_2.p2_a import closest_pair as closest_pair_a
from problems.problem_2.p2_b import closest_pair as closest_pair_b
from problems.problem_2.p2_c import closest_pair as closest_pair_c
```

```python
## Get run times 
a_times = []
b_times = []
c_times = []

# List sizes
sizes = [int(i) for i in np.linspace(0, 100, 10)]

# Get run times
for size in tqdm(sizes):

    a_times_single_size = []
    b_times_single_size = []
    c_times_single_size = []

    for _ in range(10):
        points = [[random.random(),random.random()]for i in range(size)]
        # a
        start_time = time.time()
        closest_pair_a(points)
        a_times_single_size.append(time.time() - start_time)
        
        # b
        start_time = time.time()
        closest_pair_b(points)
        b_times_single_size.append(time.time() - start_time)
        
        # c
        start_time = time.time()
        closest_pair_c(points)
        c_times_single_size.append(time.time()  - start_time)

    a_times.append(statistics.median(a_times_single_size))
    b_times.append(statistics.median(b_times_single_size))
    c_times.append(statistics.median(c_times_single_size))
```

```python
# Plotting
plt.plot(sizes, a_times, marker='o', label='$O(n^2)$', color='blue')
plt.plot(sizes, b_times, marker='o', label='$O(n \log n)$', color='red')
plt.plot(sizes, c_times, marker='o', label='$O(n)$', color='yellow')
plt.xlabel('List Size')
plt.ylabel('Time (seconds)')
plt.title('Problem 2: Empirical Performance Analysis')
plt.legend()
plt.savefig(os.path.dirname(os.getcwd()) + '/latex/figures/problem-2.png')
```

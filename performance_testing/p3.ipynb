```python
import matplotlib.pyplot as plt
import math
import numpy as np
import os
import random
import sys
import statistics
import time
from tqdm import tqdm
sys.path.append(os.path.dirname(os.getcwd()))
import problems.problem_3.p3_a as p3_a
import problems.problem_3.p3_b as p3_b
```

# Insertion

```python
## Get run times 
a_times = []
b_times = []

# List sizes
sizes = [int(i) for i in np.linspace(2, 1000, 20)]

# Get run times
for size in tqdm(sizes):

    a_times_single_size = []
    b_times_single_size = []

    for _ in range(10):
            
        ints_lst = [random.randint(0,size) for i in range(size)]
        
        # a
        start_time = time.time()
        l = p3_a.LinkedList()
        for i in ints_lst:
            l.insert(i)
        a_times_single_size.append(time.time() - start_time)
        
        # b
        max_level = 0
        if int(math.log(size)) < 2:
            max_level = 2
        else:
            max_level = int(math.log(size))
        p = random.random()
        start_time = time.time()
        l = p3_b.SkipList(max_level, p)
        for i in ints_lst:
            l.insert(i)
        b_times_single_size.append(time.time() - start_time)
        

    a_times.append(statistics.median(a_times_single_size))
    b_times.append(statistics.median(b_times_single_size))
```

```python
# Plotting
plt.plot(sizes, a_times, marker='o', label='$O(n)$', color='blue')
plt.plot(sizes, b_times, marker='o', label='$O(n \log n)$', color='red')
plt.xlabel('List Size')
plt.ylabel('Time (seconds)')
plt.title('Problem 3b: Empirical Performance Analysis Insertion')
plt.legend()
plt.savefig(os.path.dirname(os.getcwd()) + '/latex/figures/problem-3b.png')
```

```python
## Get run times 
a_times = []
b_times = []

# List sizes
sizes = [int(i) for i in np.linspace(2, 1000, 20)]

# Get run times
for size in tqdm(sizes):

    a_times_single_size = []
    b_times_single_size = []

    for _ in range(10):
            
        ints_lst = [random.randint(0,size) for i in range(size)]
        delete_lst = random.sample(ints_lst, size // 2)
        # a
        l = p3_a.LinkedList()
        for i in ints_lst:
            l.insert(i)
        start_time = time.time()
        for i in delete_lst:
            l.delete(i)
        a_times_single_size.append(time.time() - start_time)
        
        # b
        max_level = 0
        if int(math.log(size)) < 2:
            max_level = 2
        else:
            max_level = int(math.log(size))
        p = random.random()
        
        
        l = p3_b.SkipList(max_level, p)
        for i in ints_lst:
            l.insert(i)
        start_time = time.time()
        for i in delete_lst:
            l.delete(i)
        b_times_single_size.append(time.time() - start_time)
        

    a_times.append(statistics.median(a_times_single_size))
    b_times.append(statistics.median(b_times_single_size))
```

```python
# Plotting
plt.plot(sizes, a_times, marker='o', label='$O(n)$', color='blue')
plt.plot(sizes, b_times, marker='o', label='$O(n \log n)$', color='red')
plt.xlabel('List Size')
plt.ylabel('Time (seconds)')
plt.title('Problem 3b: Empirical Performance Analysis Deletion')
plt.legend()
plt.savefig(os.path.dirname(os.getcwd()) + '/latex/figures/problem-3c.png')
```
